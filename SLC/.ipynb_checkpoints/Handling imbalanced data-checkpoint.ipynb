```python
from sklearn.preprocessing import StandardScaler

df2 = df.loc[:,['Amount','Time']]
cols = df2.columns
ss = StandardScaler()

df2 = ss.fit_transform(df2)

df2 = pd.DataFrame(df2,columns=cols)


df = df.drop(['Amount','Time'],axis=1)

df = pd.concat([df,df2],axis=1)
```

```python
df.head()
```

```python
y = df['Class']
X = df.drop('Class',axis=1)
```

```python
X = X.drop(['V17','V12','V14','V10','V7','V16','V11','V3','V4','V18','V2','V1','V5','V9','V23','V6'],axis=1)
```

```python
from sklearn.metrics import accuracy_score,roc_auc_score,roc_curve
```

```python
def model_compare(algo,xtrain,xtest,ytrain,ytest):
    algo.fit(xtrain,ytrain)
    y_pred_train = algo.predict(xtrain)
    y_prob_train = algo.predict_proba(xtrain)
    
    y_pred_test = algo.predict(xtest)
    y_prob_test = algo.predict_proba(xtest)
    
    print('Accuracy on Train set', accuracy_score(ytrain,y_pred_train))
    print('Accuracy on Test set', accuracy_score(ytest,y_pred_test))
    
    print('Accuracy on Train set', roc_auc_score(ytrain,y_prob_train))
    print('Accuracy on Test set', roc_auc_score(ytest,y_prob_test))
    
    fpr,tpr,threshold = roc_curve(ytrain,y_prob,train)
    plt.plot(fpr,tpr)
    plt.plot(fpr,fpr,'r-')
    plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
```

```python
model_compare(rfc,X_train,X_test,y_train,y_test)
```
