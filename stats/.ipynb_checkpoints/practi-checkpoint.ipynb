```python
from    scipy.stats             import  ttest_1samp,ttest_ind, wilcoxon, ttest_ind_from_stats
import  scipy.stats             as      stats  
from    statsmodels.stats.power import  ttest_power
import  matplotlib.pyplot       as      plt
```

```python
income_car_yes=df[dum1['car_YES']==1]['income']
income_car_no=df[dum1['car_YES']==0]['income']
t_statistic, p_value  =  stats.ttest_ind(income_car_yes,income_car_no)
print('P Value %1.3f' % p_value)
print('t-stat Value %1.3f' % t_statistic) 
```

```python
r1=df[(df['region']=='INNER_CITY')]['income']
r2=df[(df['region']=='TOWN')]['income']
r3=df[(df['region']=='RURAL')]['income']
r4=df[(df['region']=='SUBURBAN')]['income']
stats.f_oneway(r1,r2,r3,r4)
```

as p value is greater than 0.05 we are accepting null hypothesis i.e salary is same in all the regions
