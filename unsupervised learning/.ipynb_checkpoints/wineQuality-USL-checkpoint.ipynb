```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('winequality-red.csv')
```

```python
df.info()
```

```python
df2=df.loc[:,'fixed_acidity':'alcohol']
```

```python
df2
```

```python
sns.pairplot(df2,diag_kind='kde')
```

```python
from scipy.stats import *
df_scaled=df2.apply(zscore)
```

```python
df_scaled
```

```python
from sklearn.cluster import KMeans
```

```python
cluster_range=range(1,15)
cluster_errors=[]
for num_clusters in cluster_range:
    clusters=KMeans(num_clusters,n_init=10)
    clusters.fit(df_scaled)
    labels=clusters.labels_
    print(labels,len(labels))
    centroid=clusters.cluster_centers_
    print(centroid,len(centroid))
    cluster_errors.append(clusters.inertia_)
    print(cluster_errors,len(cluster_errors))
cluster_df=pd.DataFrame({'num_clusters':cluster_range,'cluster_errors':cluster_errors})
cluster_df[0:15]
```

```python
#scree plot or elbow plot

plt.figure(figsize=(10,10))
plt.plot(cluster_df.num_clusters,cluster_df.cluster_errors,marker='*')
```

```python
Kmeans=KMeans(n_clusters=6,n_init=15,random_state=2345)
```
