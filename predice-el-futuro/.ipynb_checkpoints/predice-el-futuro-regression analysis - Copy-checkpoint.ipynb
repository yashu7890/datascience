```python
params={'n_estimators':sp_randint(5,50),
            'max_depth':sp_randint(2,30),
            'min_samples_split':sp_randint(2,20),
            'min_samples_leaf':sp_randint(1,20),
            'max_features':sp_randint(1,2)}

rand_search_rfr=RandomizedSearchCV(rfr,param_distributions=params,cv=3,random_state=1)

rand_search_rfr.fit(X_train,y_train)
rand_search_rfr.best_params_
rfr_hyp=RandomForestRegressor(**rand_search_rfr.best_params_)
rfr_hyp.fit(X_train,y_train)

y_pred_train=rfr_hyp.predict(X_train)
y_pred=rfr_hyp.predict(X_test)

```

```python
X=df.drop(['feature','time','f1'],1)
y=df['f1']
```

```python
tscv = TimeSeriesSplit()
print(tscv)

X=np.array(X)

for train_index, test_index in tscv.split(X):
    print("TRAIN:", train_index, "TEST:", test_index)
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    

    rfr.fit(X_train,y_train)

    y_pred_train=rfr.predict(X_train)
    y_pred=rfr.predict(X_test)
    

    y_pred_train=rfr.predict(X_train)
    y_pred=rfr.predict(X_test)
    
    y_train,y_pred_train=[np.e**i for i in y_train],[np.e**i for i in y_pred_train]
    y_test,y_pred=[np.e**i for i in y_test],[np.e**i for i in y_pred]
    
    print('RMSE of train:',np.sqrt(mean_squared_error(y_train,y_pred_train)))
    print('RMSE of test:',np.sqrt(mean_squared_error(y_test,y_pred)))
    
    t1=np.hstack([y_pred_train,y_pred])
    t2=np.hstack([y_train,y_test])
    t3=[]
    for i,j in X_train:
        t3.append(i*60+j)
        
    for i,j in X_test:
        t3.append(i*60+j)
    
    sns.lineplot(x=t3,y=t2)
    sns.lineplot(x=t3,y=t1)
    locs, labels = plt.xticks()
    plt.xticks(locs,time_modified,rotation=45)
    


    r2_score_train=(r2_score(y_train,y_pred_train))
    r2_score_test=(r2_score(y_test,y_pred))
    print(r2_score_train,r2_score_test)
    plt.show()

    
```
