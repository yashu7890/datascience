```python
df['time']= pd.to_datetime(df['time']) 
df['time']
```

df['time_year'] = df['time'].dt.year

```python
df['time_minute'] = df['time'].dt.minute
df['time_second'] = df['time'].dt.second
```

```python
X=df.drop(['feature','time','f1','f2'],1)
y=df['f2']
```

```python
#simple mean model....y_pred is the mean of y_train . This to compare with the model that we have generated above

tscv = TimeSeriesSplit()
print(tscv)
rfr=RandomForestRegressor(random_state=1)
X=np.array(X)

for train_index, test_index in tscv.split(X):
    print("TRAIN:", train_index, "TEST:", test_index)
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    

    rfr.fit(X_train,y_train)

    y_pred_train=rfr.predict(X_train)
    
    mea=np.mean(y_train)
    y_pred=[mea for i in y_test]
    
    print('RMSE of train:',np.sqrt(mean_squared_error(y_train,y_pred_train)))
    print('RMSE of test:',np.sqrt(mean_squared_error(y_test,y_pred)))
    
    t1=np.hstack([y_pred_train,y_pred])
    t2=np.hstack([y_train,y_test])
    t3=[]
    for i,j in X_train:
        t3.append(i*60+j)
    for i,j in X_test:
        t3.append(i*60+j)
    
    sns.lineplot(x=t3,y=t2)
    sns.lineplot(x=t3,y=t1)
    plt.xticks(rotation=45)
    


    r2_score_train=(r2_score(y_train,y_pred_train))
    r2_score_test=(r2_score(y_test,y_pred))
    print('R2 train:',r2_score_train,'R2 test:',r2_score_test)
    plt.show()

    
```
